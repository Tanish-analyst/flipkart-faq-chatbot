```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Reading file
with open("/content/formatted_flipkart_faq2.txt", "r", encoding="utf-8") as f:
    content = f.read()
print(content)
```

```python
# Splitting using "# tanish" as the delimiter
faq_chunks = content.split("# tanish")

# Clean up and rejoin for LangChain processing
faq_chunks = [chunk.strip() for chunk in faq_chunks if chunk.strip()]

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=100
)
docs = splitter.create_documents(faq_chunks)

print(f"Total chunks: {len(docs)}")
```

```python
# Printing each chunk
for i, doc in enumerate(docs):
    print(f"\n--- Chunk {i+1} ---\n")
    print(doc.page_content)
```

```python
from langchain.embeddings import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(openai_api_key="sk-proj-............")
```

```python
from langchain.vectorstores import FAISS            # for efficient similarity search
vectorstore = FAISS.from_documents(docs, embeddings)    # Converting test documents into numerical vectors
```

```python
vectorstore.index_to_docstore_id
```

```python
# Saves the created vector database (index & documents) to the specified folder for later use, avoiding the need to recompute it.
vectorstore.save_local("flipkart_faq_index")
```

```python
# Creates a "retriever" tool that fetches the top 1 most relevant document chunk from your knowledge base for any query.
retriever = vectorstore.as_retriever(search_kwargs={"k": 2})
```

```python
# checking how the retrieval works by finding relevant FAQ sections for the query
r_docs = retriever.get_relevant_documents("I order new mobile phone yesterday, how can I track my order?")
for doc in r_docs:
    print(doc.page_content)
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
template = """Answer the question based only on the following context:
{context}

Question: {question}
"""
prompt = ChatPromptTemplate.from_template(template)
```

```python
from langchain_core.runnables import RunnableLambda
retrieve_docs = RunnableLambda(retriever.invoke)
```

```python
#Creating the full chain
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

rag_chain = (
    {"context": retrieve_docs, "question": RunnablePassthrough()}
    | prompt
    | ChatOpenAI(model="gpt-3.5-turbo", openai_api_key="sk-proj-.............")
)
```

```python
def ask(question):
    return rag_chain.invoke(question)


print("🤖 Chatbot is ready with! Type 'exit' to quit.\n")

while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("👋 Goodbye!")
        break
    print("Bot:", ask(user_input))
```
